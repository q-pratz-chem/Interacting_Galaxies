Ellipse equation:$\quad \quad \quad \frac{x^2}{a^2} +\frac{y^2}{b^2} =1,$  
    where $a$ is _semi-major_ axis, $b$ is _semi-minor_ axis.

The period $T$ of the orbit in terms of $a$ by Kelpler's 3rd law:
$$a = \bigg(GM\frac{T^2}{4\pi^2}\bigg)^{\frac{1}{3}}$$

In the above figure $l$ is _semi-latus rectum_, which we define here by $p$, such that: $p = a(1-e^2)$ 

Gravitational force between two bodies,  $r$ distance apart, with masses $M$ and $m$.

$$F = \frac{GMm}{r^2} \hat{r} = \frac{GMm}{r^3} \overrightarrow{r}$$

$$F_x = \frac{GMm}{r^3}x, \quad F_y = \frac{GMm}{r^3}y $$

$$ \implies \ddot{x} = \frac{GM}{r^3}x, \quad \ddot{y} = \frac{GM}{r^3}y $$

$$ \implies \ddot{x} = \frac{GM}{{\sqrt{(x^2 + y^2)}}^3}x, \quad \ddot{y} = \frac{GM}{{\sqrt{(x^2 + y^2)}}^3}y $$

$\begin{split} \text{Acceleration} \quad f(x,y)  &:=  \ddot{x}, \ddot{y}\\ f(x_0, y_0) &= \frac{GM}{{r_0}^3}x_0, \frac{GM}{r_0^3}y_0, \quad \end{split}$ where $r_0 = \sqrt{(x_0^2 + y_0^2)}$
